# 01 — EDA: Home Credit Default Risk

Tarea académica 1 (`prd.md` §5). Exploración de las 8 tablas del dataset Home Credit Default Risk.

**Tabla central de esta primera sección:** `application_train.csv` (307,511 filas × 122 columnas esperadas).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 100)
sns.set_theme(style="whitegrid")

DATA_DIR = "../data/raw"
```

```python
app_train = pd.read_csv(f"{DATA_DIR}/application_train.csv")
print(app_train.shape)
app_train.head()
```

## Distribución de la variable objetivo

`TARGET` = 1 indica incumplimiento (default). El PRD reporta una tasa de default del 8.07% (§2.2), lo que convierte esto en un problema de clasificación binaria desbalanceada.

```python
target_counts = app_train["TARGET"].value_counts()
target_rate = app_train["TARGET"].mean()
print(target_counts)
print(f"Tasa de default: {target_rate:.4%}")

sns.countplot(x="TARGET", data=app_train)
plt.title("Distribución de TARGET (0 = paga, 1 = default)")
plt.show()
```

## Valores faltantes

Muchas columnas de `application_train` (especialmente las de datos externos y del inmueble) tienen alto porcentaje de nulos. Esto va a condicionar la estrategia de imputación en el feature engineering (Tarea 2).

```python
missing = app_train.isnull().sum()
missing_pct = (missing / len(app_train) * 100).round(2)

missing_summary = (
    pd.DataFrame({"n_missing": missing, "pct_missing": missing_pct})
    .query("n_missing > 0")
    .sort_values("pct_missing", ascending=False)
)

print(f"Columnas con al menos un nulo: {len(missing_summary)} de {app_train.shape[1]}")
missing_summary.head(20)
```

## Correlación de EXT_SOURCE_1/2/3 con TARGET

El PRD (§7.2) señala a `EXT_SOURCE_1`, `EXT_SOURCE_2` y `EXT_SOURCE_3` (scores externos normalizados) como las features más predictivas del dataset, con correlaciones reportadas de -0.160 (EXT_SOURCE_2) y -0.156 (EXT_SOURCE_3). Confirmamos con los datos propios.

```python
ext_source_cols = ["EXT_SOURCE_1", "EXT_SOURCE_2", "EXT_SOURCE_3"]
correlations = app_train[ext_source_cols + ["TARGET"]].corr()["TARGET"].drop("TARGET")
print(correlations.sort_values())

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, col in zip(axes, ext_source_cols):
    sns.kdeplot(data=app_train, x=col, hue="TARGET", ax=ax, common_norm=False)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

## Edad (DAYS_BIRTH) y default

`DAYS_BIRTH` está en la misma escala que `DAYS_EMPLOYED` (días negativos desde la solicitud). La convertimos a años y miramos si la edad se relaciona con el default.

```python
age_years = -app_train["DAYS_BIRTH"] / 365.25
print(age_years.describe())

fig, ax = plt.subplots(figsize=(8, 4))
sns.kdeplot(data=pd.DataFrame({"age": age_years, "TARGET": app_train["TARGET"]}),
            x="age", hue="TARGET", common_norm=False, ax=ax)
ax.set_title("Distribución de edad por TARGET")
ax.set_xlabel("Edad (años)")
plt.show()
```

```python
np.corrcoef(age_years, app_train["TARGET"])[0, 1]
```

**Conclusión:** correlación de -0.0782 entre edad y `TARGET` — existe relación real (a menor edad, mayor probabilidad de default) pero es notablemente más débil que la de `EXT_SOURCE_*` (-0.16/-0.18). Consistente con la lectura visual del KDE: las distribuciones se solapan bastante, con el default levemente corrido hacia edades más jóvenes.

## Variables categóricas vs. tasa de default

Miramos cómo varía la tasa de default por `CODE_GENDER`, `NAME_EDUCATION_TYPE` y `NAME_INCOME_TYPE`. Estas variables (o proxies de ellas) son relevantes más adelante para la auditoría de fairness (RNF-2), así que interesa ver si hay diferencias marcadas entre grupos ya en esta etapa exploratoria.

```python
categorical_cols = ["CODE_GENDER", "NAME_EDUCATION_TYPE", "NAME_INCOME_TYPE"]

for col in categorical_cols:
    rate_by_group = (
        app_train.groupby(col)["TARGET"]
        .agg(["mean", "count"])
        .rename(columns={"mean": "default_rate", "count": "n"})
        .sort_values("default_rate", ascending=False)
    )
    print(f"\n--- {col} ---")
    print(rate_by_group)
```

## Pasada liviana por las 7 tablas restantes

`application_train`/`application_test` tienen una fila por solicitud (`SK_ID_CURR`, clave única). Las siguientes 6 tablas son **relacionales**: registran historial (bureau externo, créditos previos con Home Credit, cuotas, saldos mensuales) y tienen múltiples filas por `SK_ID_CURR` (o por `SK_ID_PREV`, que a su vez referencia `SK_ID_CURR`). No las exploramos en profundidad acá — eso es trabajo de la Tarea 2 (feature engineering, agregación) — pero relevamos shape, claves y nulos a alto nivel.

```python
other_tables = [
    "application_test.csv",
    "bureau.csv",
    "bureau_balance.csv",
    "previous_application.csv",
    "POS_CASH_balance.csv",
    "credit_card_balance.csv",
    "installments_payments.csv",
]

summary_rows = []
for fname in other_tables:
    df = pd.read_csv(f"{DATA_DIR}/{fname}", nrows=None)
    key_col = "SK_ID_CURR" if "SK_ID_CURR" in df.columns else df.columns[0]
    summary_rows.append({
        "tabla": fname,
        "filas": df.shape[0],
        "columnas": df.shape[1],
        "ids_unicos": df[key_col].nunique() if key_col in df.columns else None,
        "clave": key_col,
    })
    del df

pd.DataFrame(summary_rows)
```

**Conclusión:** las shapes coinciden exactamente con lo documentado en `prd.md` §7.2 (688 MB comprimidos, ~2.7 GB descomprimidos). Los `ids_unicos` confirman la naturaleza relacional: por ejemplo `bureau.csv` tiene 1,716,428 filas para solo 305,811 `SK_ID_CURR` (~5.6 registros de bureau por solicitante en promedio), e `installments_payments.csv` tiene 13.6M filas para 339,587 solicitantes (~40 cuotas por solicitante en promedio). `bureau_balance.csv` se relaciona con `bureau.csv` por `SK_ID_BUREAU`, no directamente con `application_train` — es una tabla de segundo grado (`bureau_balance` → `bureau` → `application`). Esto confirma que el feature engineering (Tarea 2) requiere una estrategia de agregación por tabla antes de poder unir todo a nivel `SK_ID_CURR`.
